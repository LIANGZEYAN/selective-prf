```python
import pandas as pd

print("="*60)
print("RENAMING COLUMNS FOR DATABASE COMPATIBILITY")
print("="*60)

# 读取现有文件
print("\nReading result_interleaved_with_text.csv...")
result = pd.read_csv('result_interleaved_with_text.csv')

print(f"Loaded {len(result)} rows")
print(f"Current columns: {list(result.columns)}")

# 重命名列
result = result.rename(columns={
    'query_text': 'query',
    'passage_text': 'text'
})

print(f"New columns: {list(result.columns)}")

# 保存（覆盖原文件）
result.to_csv('result_interleaved_with_text.csv', index=False)

print("\n✓ Successfully renamed columns!")
print("✓ Saved to: result_interleaved_with_text.csv")

# 验证
print("\n" + "="*60)
print("VERIFICATION:")
print("="*60)
print(f"Total rows: {len(result)}")
print(f"Columns: {list(result.columns)}")
print(f"Required columns present: {all(col in result.columns for col in ['qid', 'query', 'docno', 'text'])}")

# 显示样例
print("\nSample data:")
print(result[['qid', 'docno', 'query', 'text']].head(2).to_string())

print("\n" + "="*60)
print("✅ DONE!")
print("="*60)
```
